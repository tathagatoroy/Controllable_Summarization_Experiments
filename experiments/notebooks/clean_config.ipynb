```python
def find_safetensors_dirs(base_path):
    safetensors_dirs = set()

    for root, dirs, files in os.walk(base_path):
        for file in files:
            if file.endswith(".safetensors"):
                safetensors_dirs.add(root)
                break  # No need to check more files in this directory

    return safetensors_dirs

# Specify the directory path

# Find directories containing .safetensors files
safetensors_dirs = find_safetensors_dirs(directory)


# Print the results
# for dir in safetensors_dirs:
#     print(dir)
#subract base directory from all subdirectories and then flatten the directory by replacing / with _
output_files = [x.replace(directory,"").replace("/","_")[1:] for x in safetensors_dirs]
attributes = [os.path.basename(x) for x in safetensors_dirs]
for output_file, dir, attribute in zip(output_files, safetensors_dirs, attributes):
    print(output_file)
    print(dir)
    print(attribute)
    print(".....")
    #run the command
```
